### Quick XGBoost Regressor Model for understanding

* specifically call out using our GPU !!! 
* predictor will be GPU based as well
* validate_parameters True to notify us if something isn't right 


```python
from numpy import asarray
import pandas as pd
from xgboost import XGBRegressor

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv'
dataframe = pd.read_csv(url, header=None)
data = dataframe.values

X, y = data[:, :-1], data[:, -1]

model = XGBRegressor()# tree_method='gpu_hist', 
                     # gpu_id=0, 
                     # verbosity=3, 
                     # predictor='gpu_predictor',
                     # validate_parameters=True)

print('\nWe will be able to see the default hyps:')
model.fit(X, y)

row = [0.00632,18.00,2.310,0,0.5380,6.5750,65.20,4.0900,
       1,296.0,15.30,396.90,4.98]
new_data = asarray([row])

yhat = model.predict(new_data)

print('Predicted: %.3f' % yhat)

print('\n')

print('\n1. Here is what we really care about:\n')

for k,v in model.get_xgb_params().items():
    print(k, "\t\t",v)
    
```
