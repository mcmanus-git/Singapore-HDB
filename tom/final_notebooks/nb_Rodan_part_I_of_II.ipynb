````python

model_features = features_to_use_for_baseline_testing_with_random_forest_regressor
target = ['price_per_sq_m_per_lease_yr_norm'] # predict this ! 
X_train =  df_train[model_features].copy()
X_train = X_train.drop(columns = 'price_per_sq_m_per_lease_yr_norm')
X_test =  df_test[model_features].copy()
X_test =  X_test.drop(columns = 'price_per_sq_m_per_lease_yr_norm')
X_val =  df_val[model_features].copy()
X_val = X_val.drop(columns = 'price_per_sq_m_per_lease_yr_norm') 
X_test_combined = pd.concat([X_test,X_val],ignore_index=False).reset_index(drop=True)
y_train =  df_train[target]
y_val   =  df_val[target]
y_test  =  df_test[target]
y_test_combined = pd.concat([y_test,y_val],ignore_index=False).reset_index(drop=True)    
y_train =  y_train.values.ravel()
y_val   =  y_val.values.ravel()
y_test  =  y_test.values.ravel()
````

````python

def RFR_model_base(X_train,y_train,X_val,y_val,X_test,y_test,X_test_combined, y_test_combined): 
    start_time = time.time()
    
    regr = RandomForestRegressor(random_state=42, 
                                 n_jobs=-1)
    
    regr.fit(X_train, y_train)
    y_pred_train =  regr.predict(X_train)
    y_pred_test  =  regr.predict(X_test)
    y_pred_val   =  regr.predict(X_val)
    y_pred_test_combined = regr.predict(X_test_combined)
    print('\n--- Test Set ---')
    print('Mean Absolute Error: ... %.5f' % mean_absolute_error(y_test, y_pred_test))
    print("Mean Squared Error:..... %.5f" % mean_squared_error(y_test, y_pred_test))
    print('RMSE: .................. %.5f' % np.sqrt(mean_squared_error(y_test, y_pred_test)))   
    print("Coeff of det (R^2):..... %.5f" % r2_score(y_test, y_pred_test))  
    print('\n--- Val Set ---')
    print('Mean Absolute Error: ... %.5f' % mean_absolute_error(y_val, y_pred_val))
    print("Mean Squared Error:..... %.5f" % mean_squared_error(y_val, y_pred_val))
    print('RMSE: .................. %.5f' % np.sqrt(mean_squared_error(y_val, y_pred_val)))   
    print("Coeff of det (R^2):..... %.5f" % r2_score(y_val, y_pred_val))
    print('\n--- Train Set ---')
    print('Mean Absolute Error: ... %.5f' % mean_absolute_error(y_train, y_pred_train))
    print("Mean Squared Error:..... %.5f" % mean_squared_error(y_train, y_pred_train))
    print('RMSE: .................. %.5f' % np.sqrt(mean_squared_error(y_train, y_pred_train)))
    print("Coeff of det (R^2):..... %.5f" % r2_score(y_train, y_pred_train)) 
    
    print('\n--- Combined Test Set ---')
    print('Mean Absolute Error: ... %.5f' % mean_absolute_error(y_test_combined, y_pred_test_combined))
    print("Mean Squared Error:..... %.5f" % mean_squared_error(y_test_combined, y_pred_test_combined))
    print('RMSE: .................. %.5f' % np.sqrt(mean_squared_error(y_test_combined, y_pred_test_combined)))
    print("Coeff of det (R^2):..... %.5f" % r2_score(y_test_combined, y_pred_test_combined))   
    print('\nModel Hyperparameters Definitions:')
    # pprint(regr.get_params(deep=True))
    for k,v in regr.get_params(deep=True).items():
        print(k, "->", v)
    print("\n--- %s seconds ---" % round((time.time() - start_time)))
          
````
