### Example 1 

*first table from this pdf looks like this:*



````python

pdf_path = "https://github.com/chezou/tabula-py/raw/master/tests/resources/data.pdf"
dfs = tabula.read_pdf(pdf_path, stream=True)
# read_pdf returns list of DataFrames
print(len(dfs))
dfs[0]

# 'pages' argument isn't specified, so will extract only from page 1 by default ! 
````

```Notes: As of tabula-py 2.0.0, read_pdf() sets multiple_tables=True by default. If you want to get consistent
output with previous version, set multiple_tables=False.```

````python

# set pages option

dfs = tabula.read_pdf(pdf_path, pages=3, stream=True)

dfs[0]

# remember to do len(dfs) to see how many mini-tables you have within it, if you are 
# extracting page by page... 
````

````python

# extract all pages

# extract all pages

tabula.read_pdf(pdf_path, pages="all", stream=True)
````

````python

# IF YOU WANTED EACH TABLE/PAGE TO BE ITS OWN DF...

stuart_pdf = "https://www.moe.gov.sg/microsites/psle-fsbb/assets/infographics/transition-to-secondary-school/indicative-psle-score-ranges-for-individual-secondary-schools.pdf"

dfs = tabula.read_pdf(stuart_pdf, pages="all")

len(dfs)

# this is assuming each of the 16 pages was its own independent table... which in this case its not, but this would 
# have 16 dfs within the dfs
````

````python

# now lets just put all tables into one large super table / df (all 16 pages)

# To read multiple tables we need to add extra parameter
#   multiple_tables = True   ->  Read multiple tables as independent tables
#   multiple_tables = False  ->  Read multiple tables as single table


stuart_pdf = "https://www.moe.gov.sg/microsites/psle-fsbb/assets/infographics/transition-to-secondary-school/indicative-psle-score-ranges-for-individual-secondary-schools.pdf"

df = tabula.read_pdf(stuart_pdf, pages="all", guess=True, multiple_tables=False)

df = df[0]  # its one big table

df.head(50)
````
