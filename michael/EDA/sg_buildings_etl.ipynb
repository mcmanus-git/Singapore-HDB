```python
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine
from MyCreds.mycreds import Capstone_AWS_PG
from geoalchemy2 import Geometry
import shapely
```

```python
query = 'select * from sg_buildings_postal_geo;'

engine = create_engine(f'postgresql+psycopg2://{Capstone_AWS_PG.username}:{Capstone_AWS_PG.password}@{Capstone_AWS_PG.host}/capstone', echo=False)

sql_alc_cnxn = engine.connect()
postal = gpd.read_postgis(query, sql_alc_cnxn, geom_col='geometry')
postal
```

```python
postal = postal.to_crs(4326)
postal
```

```python
postal['geometry'] = postal['geometry'].map(lambda point: shapely.ops.transform(lambda x, y: (y, x), point))
postal
```

```python
engine = create_engine(f'postgresql+psycopg2://{Capstone_AWS_PG.username}:{Capstone_AWS_PG.password}@{Capstone_AWS_PG.host}/capstone', echo=False)

sql_alc_cnxn = engine.connect()

postal.to_postgis('sg_buildings_postal_geo', sql_alc_cnxn, if_exists='replace', index=False, dtype={'geometry': Geometry(geometry_type='POINT', srid=4326)})

sql_alc_cnxn.close()
```
