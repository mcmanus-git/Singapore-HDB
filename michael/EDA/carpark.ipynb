```python
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine
from MyCreds.mycreds import Capstone_AWS_PG
from geoalchemy2 import Geometry
import shapely
```

```python
carpark = gpd.read_file('../data/hdb-carpark-information/hdb-carpark-information.csv')
carpark
```

```python
carpark['geometry'] = gpd.points_from_xy(carpark['y_coord'], carpark['x_coord'], crs='EPSG:3414')
carpark
```

```python
carpark = carpark.to_crs(4326)
carpark
```

```python
carpark['geometry'] = carpark['geometry'].map(lambda point: shapely.ops.transform(lambda x, y: (y, x), point))
carpark
```

```python
engine = create_engine(f'postgresql+psycopg2://{Capstone_AWS_PG.username}:{Capstone_AWS_PG.password}@{Capstone_AWS_PG.host}/capstone', echo=False)

sql_alc_cnxn = engine.connect()

carpark.to_postgis('hdb_carpark_info', sql_alc_cnxn, if_exists='replace', index=False, dtype={'geometry': Geometry(geometry_type='POINT', srid=4326)})

sql_alc_cnxn.close()
```
